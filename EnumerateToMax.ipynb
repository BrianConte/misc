```python
max = [3,4,1,3,1]
comb = [0 for j in range(len(max))]
iEnd = 0
while iEnd < len(max):
    print[comb]
    while comb[iEnd] == max[iEnd] and iEnd < len(max):
        iEnd +=1 
        if iEnd >= len(max):
            break
    for iStart in range(iEnd):
        comb[iStart]=0
    if iEnd < len(max):
        comb[iEnd]+=1
        iEnd = 0
```
